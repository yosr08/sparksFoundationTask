```python
import nltk
nltk.download('vader_lexicon')
```

```python
#import VADER sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#creating an instance
vds = SentimentIntensityAnalyzer()
```

```python
#sentiment analysis of the below text data
text = 'I went to the movie, yesterday. It was amazing! Everyone acted well'
#polarity_scores followed by the text data inside
print('n-> Sentiment scores of the text data:\n')
vds.polarity_scores(text)
```

* Positive sentiment: compound score >= 0.05
* Neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
* Negative sentiment: compound score <= -0.05

```python
import pandas as pd
```

```python
data_news = pd.read_csv("india-news-headlines.csv" , parse_dates=[0] )
```

```python
data_news.head()
```

```python
data_news['headline_text'] = data_news[['publish_date','headline_text']].groupby(['publish_date'])['headline_text'].transform(lambda x: ','.join(x))
```

```python
data_news.drop("headline_category",axis=1,inplace=True)
```

```python
data_news.drop_duplicates(inplace=True)
```

```python
data_news.head()
```

```python
#Create an empty list to store sentiment scores
scores = []
for r in data_news.headline_text:
    scores.append(vds.polarity_scores(r))
```

```python
scores
```

```python
scores = pd.DataFrame(scores)
scores
```

```python
scores['sentiment'] = scores['compound']>=0.05
```

```python
scores
```

```python
scores.sentiment=scores.sentiment.map({False: 'negative', True:'positive'})
```

```python
scores.index = data_news.publish_date
```

```python
scores
```

```python
#export data to csv
scores.to_csv("sentiment.csv" , index=True)
```
